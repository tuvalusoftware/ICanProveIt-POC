```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ('user', '{input}')
])
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
chain = prompt | llm | output_parser
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('./C7.pdf')
pages = loader.load_and_split()
```

```python
result = chain.invoke({'input': 'Generate list of main points, format: Title (start_page-end_page)', 'context': pages})
```

```python
result.split('\n')
```

```python
import time
```

```python
questions = []

for sec in result.split('\n'):
    questions_text = chain.invoke({'input': f'Generate questions and answer of {sec}, format: q: question -- a: answer', 'context': pages})
    print(questions_text)
    q = []

    questions_split = questions_text.split('\n')

    for i in range(0, len(questions_text.split('\n')), 3):
        q.append({'question': questions_split[i].replace('q:', '').strip(), 'answer': questions_split[i+1].replace('a:', '').strip()})

    questions.append({'section': sec, 'questions': q})
    time.sleep(20)
```
